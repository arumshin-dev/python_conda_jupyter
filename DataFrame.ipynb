````python
patient_df['bmi']=round(patient_df['weight'] / patient_df['height']**2, 1)
````

````python
#5-3. 정규화
# scaling:스케일링 - 숫자 데이터의 단위나 범위를 일정하게 맞추는 것. 예: 키 m, 몸무게 kg
# 정규화(Normalization) 0 과 1 사이로 나타냄 최대:1, 최소:0
(patient_df['height'] - patient_df['height'].min()) / \
(patient_df['height'].max() - patient_df['height'].min())
````

````python
#5-3. 정규화
# scaling:스케일링 - 숫자 데이터의 단위나 범위를 일정하게 맞추는 것. 예: 키 m, 몸무게 kg
# 정규화(Normalization) 0 과 1 사이로 나타냄 최대:1, 최소:0
patient_df['height_n'] = (patient_df['height'] - patient_df['height'].min()) / \
(patient_df['height'].max() - patient_df['height'].min())
````

````python
patient_df['height_n'].describe()
````

````python
patient_df['weight_n']=(patient_df['weight'] - patient_df['weight'].min())/\
(patient_df['weight'].max()-patient_df['weight'].min())
patient_df
````

````python
def group_age(x):
    if x >= 10 and x < 20:
        return '10s'
    elif x >= 20 and x < 30:
        return '20s'
    elif x >= 30 and x < 40:
        return '30s'
    elif x >= 40 and x < 50:
        return '40s'
    elif x >= 50 and x < 60:
        return '50s'
    else:
        return '60s'
patient_df['age_group1'] = patient_df['age'].apply(group_age)
patient_df
````

````python
#함수
def multiply_two(x):
    return 2*x

#람다 함수
lambda x: 2*x

#apply()와 람다 함수를 함께 사용
#표기 단위를 미터에서 인치로 바꾸고 싶다면, 1미터는 약 39.370079인치
patient_df['height'].apply(lambda x: 39.370079*x)
#patient_df['height'] * 39.370079

#bmi라는 컬럼에 BMI 값을 계산해 두고, 
# BMI 값이 30이 넘으면 비만('Y'), 넘지 않으면 비만이 아니라고('N') 구분한다고 할게요. 
# 그러려면 콜론 뒤에 바로 조건식이 참일 때 리턴할 값인 Y를 써 주고, if문을 쓴 다음에는 else 뒤에 조건식이 거짓일 때 리턴할 값인 N을 써 주면 됩니다.
patient_df['bmi'] = round(patient_df['weight'] / patient_df['height']**2, 1)
patient_df['bmi'].apply(lambda x: 'Y' if x > 30 else 'N')

def isin_interval(x, n_lower, n_upper):
    if n_lower <= x < n_upper:
        return 'Y'
    else:
        return 'N'
# 18.5와 25를 써 주면 BMI가 18.5 이상 25 미만일 때 Y값이 나옵니다. 보통 이 구간을 정상 체중
patient_df['bmi'].apply(lambda x: isin_interval(x, 18.5, 25))
# 과체중 구간을 구하려면 아래 코드처럼 isin_interval()에 x와 25, 30을 넘겨주면 됩니다.
patient_df['bmi'].apply(lambda x: isin_interval(x, 25, 30))
patient_df
````

## 6-2. 다양한 형식의 날짜와 시간 데이터 다루기

### 1. **read_csv로 날짜/시간(datetime) 컬럼 불러오기**

- **기본 사용**  
  ```python
  pd.read_csv('파일명.csv', parse_dates=['컬럼1', '컬럼2'])
  ```
  - 날짜 형식이 `YYYY-MM-DD` (2015-01-20)처럼 pandas가 자동 인식 가능한 형식이면 바로 datetime으로 읽어옴

- **구분자/포맷이 달라도 parse_dates로 자동 변환 가능**
  - 예: `YYYY/MM/DD`, `MM/DD/YYYY` 등도 parse_dates로 대체로 자동 인식됨

- **일/월/연도(유럽식) → dayfirst=True 옵션**
  ```python
  pd.read_csv('order3.csv', parse_dates=['order_time'], dayfirst=True)
  ```

### 2. **to_datetime으로 더 다양한 날짜 형식 처리**

- **특정 포맷 지정 (예: 2자리 연도, 한글 등)**
  - 연도/월/일이지만 2자리 연도면 `yearfirst=True`
  - 다양한 포맷이면 `format` 옵션 사용
  ```python
  pd.to_datetime(df['date_col'], yearfirst=True)
  pd.to_datetime(df['date_col'], format='%Y년 %m월 %d일 %H시 %M분 %S초')
  ```

### 3. **datetime ↔ 문자열 원하는 포맷으로 바꾸기**

- **datetime 객체 → 문자열 포맷**
  ```python
  df['date_col'].dt.strftime('%d %b %Y, %I:%M %p')
  ```
  - `%d`: 일, `%b`: 월 영문 약자, `%Y`: 연도, `%I`: 12시간, `%M`: 분, `%p`: AM/PM

### 4. **시간대 처리 (타임존)**

- **시간대 없는 datetime 컬럼에 타임존 부여**  
  ```python
  df['col'] = df['col'].dt.tz_localize('Asia/Seoul')
  ```
- **이미 타임존 부여된 datetime을 다른 타임존으로 변환**  
  ```python
  df['col'] = df['col'].dt.tz_convert('America/New_York')
  ```
- **타임존 정보 출력**
  ```python
  import pytz
  for tz in pytz.all_timezones:
      print(tz)
  ```
