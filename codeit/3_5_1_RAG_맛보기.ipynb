```python
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
import getpass
import os

# Prompt the user for the OpenAI API key securely
os.environ["OPENAI_API_KEY"] = getpass.getpass("Enter your OpenAI API key: ")
```

```python
# 1. ëª¨ë¸ ì¤€ë¹„
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# 2. ìš°ë¦¬ ê°€ê²Œë§Œì˜ ì§€ì‹ (Context) - ì•„ì§ DB ì—†ì´ ë³€ìˆ˜ë¡œ ì§ì ‘ ì •ì˜
context = """
[ë§›ìˆëŠ” ê¹€ë°¥ì§‘ ë©”ë‰´íŒ]
1. ì›ì¡° ê¹€ë°¥: 3,000ì› - 40ë…„ ì „í†µì˜ ë§›
2. ì¹˜ì¦ˆ í­íƒ„ ê¹€ë°¥: 4,500ì› - ëª¨ì§œë ë¼ ì¹˜ì¦ˆê°€ ë“¬ë¿
3. ë¶ˆë‹­ ê¹€ë°¥: 5,000ì› - ì•„ì£¼ ë§¤ì›€, ë§µì°”ì´ ê¸ˆì§€
* ì˜ì—… ì‹œê°„: ì˜¤ì „ 10ì‹œ ~ ì˜¤í›„ 8ì‹œ (ë§¤ì£¼ ì›”ìš”ì¼ íœ´ë¬´)
"""
```

```python
# 3. í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿ ë§Œë“¤ê¸°
# RAGì˜ í•µì‹¬: {context} ìë¦¬ì— ê²€ìƒ‰ëœ ë¬¸ì„œë¥¼ ë„£ëŠ” ê²ƒ!
template = """
ë‹¹ì‹ ì€ ì¹œì ˆí•œ ê¹€ë°¥ì§‘ ì ì›ì…ë‹ˆë‹¤.
ì•„ë˜ [ì°¸ê³  ì •ë³´]ë¥¼ ë°”íƒ•ìœ¼ë¡œ ì†ë‹˜ì˜ [ì§ˆë¬¸]ì— ë‹µë³€í•´ì£¼ì„¸ìš”.
ì •ë³´ì— ì—†ëŠ” ë‚´ìš©ì€ "ì˜ ëª¨ë¥´ê² ìŠµë‹ˆë‹¤"ë¼ê³  ë‹µí•˜ì„¸ìš”.

[ì°¸ê³  ì •ë³´]
{context}

[ì§ˆë¬¸]: {question}
"""

prompt = PromptTemplate.from_template(template)
```

```python
# 4. LangChainì˜ íŒŒì´í”„ë¼ì¸: LangChain Expression Language(LCEL)
# í”„ë¡¬í”„íŠ¸ -> LLM -> ë¬¸ìì—´ ë³€í™˜
chain = prompt | llm | StrOutputParser()
```

```python
# 5. ì‹¤í–‰ ë° ê²°ê³¼ í™•ì¸
question = "ë¶ˆë‹­ ê¹€ë°¥ ì–¼ë§ˆì¸ê°€ìš”? ê·¸ë¦¬ê³  ì§€ê¸ˆ ì €ë… 9ì‹œì¸ë° ê°€ë„ ë˜ë‚˜ìš”?"

response = chain.invoke({"context": context, "question": question})

print("=== AI ë‹µë³€ ===")
print(response)
```
