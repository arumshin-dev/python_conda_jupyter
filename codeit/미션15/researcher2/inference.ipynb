```python
import pandas as pd
# import pickle
import joblib

# 경로
BASE_PATH = "/home/jovyan/work/shared"

# 모델 로드
# with open(MODEL_PATH, "rb") as f:
#     model = pickle.load(f)
model = joblib.load(f"{BASE_PATH}/model.pkl")

# 테스트 데이터 로드
test_df = pd.read_csv(f"{BASE_PATH}/test.csv")
# 인코딩(학습과 동일)
test_df['Extracurricular Activities'] = test_df['Extracurricular Activities'].map({
    'Yes': 1,
    'No': 0
})
# 추론
pred = model.predict(test_df.drop(columns=["Performance Index"], errors="ignore"))
# 6. 가장 가까운 정수로 반올림 + 범위 보정
pred_rounded = (
    pd.Series(pred)
    .round()
    .clip(10, 100)  # 성취도 지수 범위 보장
)
# test_df["prediction"] = pred_rounded.astype(int)

# 결과 저장
# test_df.to_csv('result.csv', index=False)
result_df = pd.DataFrame({
    "Prediction":pred_rounded
    })
result_df.to_csv('result.csv', index=False)
print("result.csv 저장 완료")
```
