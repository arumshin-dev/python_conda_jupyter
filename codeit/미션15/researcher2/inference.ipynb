```python
import pandas as pd
# import pickle
import joblib

# 경로
MODEL_PATH = "/app/output/model.pkl"
TEST_PATH = "/app/output/test.csv"
RESULT_PATH = "/app/output/result.csv"

# 모델 로드
# with open(MODEL_PATH, "rb") as f:
#     model = pickle.load(f)
model = joblib.load('shared/model.pkl')

# 테스트 데이터 로드
test_df = pd.read_csv(TEST_PATH)

# 추론
pred = model.predict(test_df.drop(columns=["target"], errors="ignore"))
test_df["prediction"] = pred

# 결과 저장
test_df.to_csv(RESULT_PATH, index=False)
# test.to_csv('shared/result.csv', index=False)
print("result.csv 저장 완료")
```
