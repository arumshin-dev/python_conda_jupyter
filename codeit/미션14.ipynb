# 문서 청킹
- RecursiveCharacterTextSplitter는 문자를 기준으로 재귀적으로 분할하는 방식.
- Character Splitters: 단순히 문자 개수를 기준으로 나눔. (CharacterTextSplitter)
- Token Splitters: 토큰(단어 또는 그 일부) 개수를 기준으로 나눔. LLM이 토큰을 기준으로 작동하기 때문에 더 정확할 수 있습니다.
- Semantic Splitters: 문서의 의미적 단락이나 문장 구조를 이해하여 의미적으로 연관된 내용을 한 청크로 묶는 방식. (예: RecursiveCharacterTextSplitter가 어느 정도 이를 고려하지만, 더 복잡한 알고리즘도 있음.)
- Document-specific Splitters: 코드, 마크다운, PDF 등 특정 문서 형식의 구조(제목, 부제목, 테이블 등)를 파악하여 청킹하는 방법. 예를 들어, PDF에서 테이블 데이터를 더 잘 처리하기 위해 UnstructuredPDFLoader를 사용한 것도 이러한 맥락이다.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
#@title RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100,
    length_function=len,
)
all_splits = text_splitter.split_documents(pages)
```

# Vector DB에 저장

1. 임베딩 모델 (Embedding Models)
- 범용 임베딩 모델: 다양한 도메인과 태스크에 사용할 수 있도록 대규모 데이터로 학습된 모델입니다.
    - nlpai-lab/KoE5 (현재 사용 중): 한국어 문장 임베딩에 특화된 모델로, 의미 유사도 측정에 좋은 성능을 보여줍니다.
    - KURE-v1: 한국어 텍스트에 적합한 임베딩 모델. nlpai-lab에서 개발한 또 다른 한국어 텍스트 특화 임베딩 모델로, 한국어 문맥 이해에 강점을 가짐.
    - all-MiniLM-L6-v2 (영어 기반): 작고 빠르면서도 성능이 좋은 영어 범용 임베딩 모델입니다.
    - OpenAI Embeddings (text-embedding-ada-002): OpenAI에서 제공하는 강력한 범용 임베딩 모델입니다.
- 도메인 특화 임베딩 모델: 특정 분야(예: 법률, 의료, 금융)의 전문 용어나 문맥을 더 잘 이해하도록 해당 분야의 데이터로 추가 학습된 모델입니다.
- 다국어 임베딩 모델: 여러 언어의 텍스트를 하나의 벡터 공간에 임베딩하여, 다른 언어 간에도 의미 유사도를 비교할 수 있게 합니다.
2. 벡터 데이터베이스 (Vector Databases)
- 인메모리 (In-memory) 벡터 DB: 메모리 내에서 작동하여 매우 빠르지만, 대규모 데이터셋에는 부적합할 수 있습니다.
    - FAISS (현재 사용 중): Facebook AI Research에서 개발한 라이브러리로, 대규모 벡터 검색에 매우 효율적입니다. Colab 환경에서 가볍게 시작하기 좋습니다.
- ChromaDB: FAISS보다 사용 편의성과 영속성이 뛰어나며, 클라우드 관리형 DB보다는 가볍게 시작할 수 있는 대안입니다.
- 스탠드얼론 (Standalone) 벡터 DB: 자체 서버나 클러스터에서 운영되며, 영속성 및 확장성을 제공합니다.
    - Weaviate
    - Pinecone
    - Qdrant
    - Milvus
- 클라우드 관리형 (Cloud-managed) 벡터 DB: 클라우드 서비스 형태로 제공되어, 인프라 관리 부담 없이 사용할 수 있습니다.
    - Amazon Aurora (벡터 확장)
    - Redis Stack (벡터 검색 모듈)

```python
import faiss
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS


# embedding_model_name = "nlpai-lab/KURE-v1"
# @title 임베딩 "nlpai-lab/KoE5", 벡터DB FAISS
embedding_model_name = "nlpai-lab/KoE5"

embeddings = HuggingFaceEmbeddings(
    model_name=embedding_model_name,
)

embedding_dim = len(embeddings.embed_query("hello world"))
index = faiss.IndexFlatL2(embedding_dim)

vector_store = FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
ids = vector_store.add_documents(documents=all_splits)

retriever = vector_store.as_retriever()
```

1. nlpai-lab/KULLM3
- 특징: KULLM3는 nlpai-lab에서 개발한 한국어 특화 LLM입니다. 보통 '한국어 능력'에 초점을 맞춰 대규모 한국어 데이터셋으로 학습되거나 추가 학습된 모델을 의미합니다. 한국어의 복잡한 문법, 어휘, 문화적 맥락을 잘 이해하고 생성하는 데 강점을 가집니다.
- 성능: 일반적으로 한국어에 대한 이해도가 높아 자연스럽고 정확한 한국어 답변을 생성하는 데 유리합니다.
- RAG에서의 역할: 검색된 한국어 맥락(Context)을 바탕으로 질문에 대한 답변을 한국어로 매끄럽게 요약하고 생성하는 데 적합합니다.

2. Bllossom/llama-3.2-Korean-Bllossom-AICA-5B
- 특징: 이 모델은 Meta의 Llama-3와 같은 강력한 오픈소스 기반 모델을 한국어 데이터셋으로 **추가 학습(fine-tuning)**하여 한국어 성능을 강화한 버전입니다. 'Bllossom'은 특정 개발 그룹이나 커뮤니티의 이름일 가능성이 높으며, 'AICA-5B'는 모델의 버전이나 크기(예: 50억 개의 파라미터)를 나타낼 수 있습니다.
- 성능: Llama-3의 강력한 기반 성능과 함께 한국어 특화 학습을 통해 영어 기반의 범용 모델보다 한국어 처리 능력이 향상되었습니다. 모델의 크기(5B)는 상대적으로 가벼운 편에 속하여 Colab과 같은 제한된 환경에서도 구동 가능성을 높입니다.
- RAG에서의 역할: 전반적인 추론 능력과 언어 생성 능력은 Llama 기반 모델에서 가져오고, 한국어에 대한 정확성과 자연스러움은 추가 학습을 통해 보강하여 RAG 시스템에서 고품질의 한국어 답변을 기대할 수 있습니다.


```python
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline

# from langchain.schema.runnable import RunnablePassthrough
# from langchain.schema.output_parser import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline
import torch

# @title llm KULLM3
language_model_name = "nlpai-lab/KULLM3"
# language_model_name = "Bllossom/llama-3.2-Korean-Bllossom-AICA-5B"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    llm_int8_enable_fp32_cpu_offload=True,
)
model = AutoModelForCausalLM.from_pretrained(
    language_model_name,
    quantization_config=bnb_config,
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained(language_model_name)
```

```python
from transformers import pipeline

llm_pipeline = pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    do_sample=True,
    temperature=0.2,
    repetition_penalty=1.2,
    return_full_text=False,
    max_new_tokens=1000,
)
llm = HuggingFacePipeline(pipeline=llm_pipeline)
chat_model = ChatHuggingFace(llm=llm)
```

# RAG

```python
from langchain_core.prompts import PromptTemplate

template = """
아래에 주어진 맥락을 이용해 질문에 대해 답변해 줘.
주어진 맥락으로 답변이 어려운 상황이라면, 그냥 모른다고 답하면 되고 억지로 답변을 꾸며 내지 마.
최대한 자세하게 답변해 줘.
반드시 한국어로 답변해야 해.

맥락:
{context}

질문:
{question}
"""

prompt = PromptTemplate.from_template(template)
```

```python
# from langchain.schema.runnable import RunnablePassthrough
# from langchain.schema.output_parser import StrOutputParser

def format_docs(docs):
    print(docs)
    return "\n\n".join(doc.page_content for doc in docs)

retrieval_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
question = "연말 정산 때 비거주자가 주의할 점을 알려 줘."

result = retrieval_chain.invoke(question)
print(result)
```

```python
# ======================================
# 1. 환경 변수 로드 (.env)
# ======================================
# from dotenv import load_dotenv
# import os

# load_dotenv()
# assert os.getenv("OPENAI_API_KEY"), "OPENAI_API_KEY가 .env에 없습니다"
import os
from getpass import getpass

# OpenAI API key
openai_api_key = getpass("Enter your OpenAI API key: ")

# API키 설정
os.environ["OPENAI_API_KEY"] = openai_api_key
```

```python

# ======================================
# 2. PDF 로딩
# ======================================
import time
from langchain_community.document_loaders import PyPDFLoader

file_path = "/content/year-end-tax.pdf"  # 🔴 실제 경로로 수정

t0 = time.time()
loader = PyPDFLoader(file_path)
docs = loader.load()
print("📄 PDF 로딩:", round(time.time() - t0, 2), "초")

# ======================================
# 3. 문서 쪼개기
# ======================================
from langchain_text_splitters import RecursiveCharacterTextSplitter

t0 = time.time()
splitter = RecursiveCharacterTextSplitter(
    chunk_size=700,
    chunk_overlap=150,
)
splits = splitter.split_documents(docs)
print("✂️ 문서 분할:", round(time.time() - t0, 2), "초")

# ======================================
# 4. 🚀 임베딩 (OpenAI API)
# ======================================
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

t0 = time.time()
embeddings = OpenAIEmbeddings(
    model="text-embedding-3-small"
)

vectorstore = FAISS.from_documents(splits, embeddings)
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
print("🧠 임베딩:", round(time.time() - t0, 2), "초")

# ======================================
# 5. LLM (OpenAI API)
# ======================================
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-5-mini",   # 속도/비용 최적
    temperature=0.2,
)

# ======================================
# 6. 프롬프트
# ======================================
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template("""
아래 [맥락]을 근거로 질문에 답하세요.
맥락에 없는 내용은 추측하지 말고 "문서에 없음"이라고 답하세요.
반드시 한국어로 답하세요.

[맥락]
{context}

[질문]
{question}
""")

# ======================================
# 7. RAG 체인
# ======================================
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

rag_chain = (
    {
        "context": retriever | format_docs,
        "question": RunnablePassthrough(),
    }
    | prompt
    | llm
    | StrOutputParser()
)

# ======================================
# 8. 테스트
# ======================================
question = "기부금 공제 시 주의사항은?"

t0 = time.time()
answer = rag_chain.invoke(question)
print("\n⏱️ 응답 시간:", round(time.time() - t0, 2), "초")
print("\n📌 답변:\n", answer)
```

```python
from langchain_core.runnables import RunnableParallel, RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
import time

# 기존 format_docs 함수는 컨텍스트 내용을 출력하는 용도였으므로,
# 체인 내부에서 사용할 새로운 함수를 정의합니다.
# 이 함수는 retriever가 반환하는 Document 객체 리스트를 받아서 LLM에 전달할 context 문자열을 생성합니다.
def format_docs_for_context(docs):
    return "\n\n".join(doc.page_content for doc in docs)

# 답변과 원본 문서를 함께 반환하는 새로운 RAG 체인을 정의합니다.
qa_chain_with_sources = (
    { # 입력으로 question을 받고, retriever를 통해 context를 가져옵니다.
        "context": retriever,  # retriever는 Document 객체 리스트를 반환합니다.
        "question": RunnablePassthrough()
    }
    | RunnableParallel(
        answer=(
            { # LLM에 전달할 최종 context와 question을 구성합니다.
                "context": lambda x: format_docs_for_context(x["context"]), # Document 리스트를 문자열로 변환
                "question": lambda x: x["question"],
            }
            | prompt # 프롬프트 템플릿 적용
            | llm      # LLM 호출
            | StrOutputParser() # LLM 출력 파싱
        ),
        source_documents=lambda x: x["context"] # retriever가 반환한 원본 Document 리스트를 그대로 전달
    )
)

# 여러 질문에 대해 테스트하고 결과를 출력하는 헬퍼 함수를 정의합니다.
def test_question_with_sources(question):
    print(f"\n--- 질문: {question} ---")
    t0 = time.time()
    result = qa_chain_with_sources.invoke(question)
    response_time = round(time.time() - t0, 2)
    print(f"⏱️ 응답 시간: {response_time} 초")
    print(f"📌 답변:\n{result['answer']}")
    print("\n[참고 문서 출처]")
    if result['source_documents']:
        for i, doc in enumerate(result['source_documents']):
            # 페이지 라벨(page_label)이 있는 경우 함께 출력
            page_info = doc.metadata.get('page_label', 'N/A')
            print(f"- 문서 {i+1} (페이지: {page_info}): {doc.page_content[:150]}...")
    else:
        print("- 참고할 문서를 찾을 수 없습니다.")
    print("=" * 50 + "\n")

# 다양한 질문으로 RAG 체인 테스트
test_question_with_sources("연말 정산 때 비거주자가 주의할 점을 알려 줘.")
test_question_with_sources("2024년 개정 세법 중에 월세와 관련한 내용이 있을까?")
test_question_with_sources("기부금 공제 때 주의할 점은?")

test_question_with_sources("오늘 저녁 메뉴 추천해 줄 수 있어?") # 문서에 없는 질문
test_question_with_sources("안녕") # 문서에 없는 질문
```
