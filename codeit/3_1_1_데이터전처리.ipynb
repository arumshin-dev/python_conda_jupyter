### íŠ¹ìˆ˜ ë¬¸ì ì œê±°í•˜ê¸°

íŠ¹ìˆ˜ ë¬¸ì ì œê±°ëŠ” íŒŒì´ì¬ ë‚´ì¥ ëª¨ë“ˆì¸ `re`ëª¨ë“ˆì„ ì‚¬ìš©í•©ë‹ˆë‹¤
> re.sub(ìˆ˜ì •ë  ë¬¸ìì—´ì˜ ì •ê·œì‹, ìˆ˜ì •í›„ì˜ ë¬¸ìì—´, ë¬¸ìì—´)

\w: Aâ€“Z, aâ€“z, 0â€“9, _

```python
### íŠ¹ìˆ˜ ë¬¸ì ì œê±°í•˜ê¸°

import re
rtext = re.sub(r'[^\w\s]', '', text)    # ìˆ«ì(\w), ê³µë°±(\s)ì„ ì œì™¸í•œ ëª¨ë“  íŠ¹ìˆ˜ë¬¸ì ì œê±°
rtext = re.sub(r'[\n\t]', ' ', rtext)   # ì¤„ë°”ê¿ˆ(\n), íƒ­(\t) ì œê±°
rtext = re.sub(r'\s+', ' ', rtext)      # ì—°ì†ëœ ê³µë°±(\s+) ì œê±°
rtext
```

### ë¶ˆìš©ì–´ ì œê±°

ë¶ˆìš©ì–´ ì œê±°ëŠ” íŒŒì´ì¬ì˜ ë¬¸ìì—´ í•¨ìˆ˜ì™€ ë¦¬ìŠ¤íŠ¸, ë°˜ë³µë¬¸ë“±ì„ ì´ìš©í•˜ì—¬ êµ¬ì„± ê°€ëŠ¥í•©ë‹ˆë‹¤.    
ë¨¼ì € í† í°í™” ì´í›„ ë°˜ë³µë¬¸ì„ í†µí•´ì„œ í•´ë‹¹ ë‹¨ì–´ê°€ ë¶ˆìš©ì–´ì— ì†í•˜ëŠ”ì§€ `if` `in` ë¬¸ì„ í™œìš©í•˜ì—¬ ì²´í¬í•˜ê³  ì†í•˜ì§€ ì•Šìœ¼ë©´ ìƒˆë¡œìš´ ë¦¬ìŠ¤íŠ¸ì— ë‹´ì•„ `join`í•´ì¤ë‹ˆë‹¤.

```python
print(rtext.split(' '))
```

```python
# í† í°í™” í›„ ë°˜ë³µë¬¸ì„ í™œìš©
sptext = rtext.split(' ')
st_words = ['', 'ì§„ì§œ', 'ì•Šì§€ë§Œ']
nw = []
for w in sptext:
  if w not in st_words:
    nw.append(w)
rtext = ' '.join(nw)
rtext
```

### ì „ì²´ ë°ì´í„°ì— í´ë Œì§• ì ìš©

DataFrame ì‚¬ìš©ì‹œ apply() ì‚¬ìš©í•˜ì—¬ ì¼ê´„ì²˜ë¦¬í•´ì¤ë‹ˆë‹¤


```python
import re

stopwords = ['', 'ì§„ì§œ','ì•Šì§€ë§Œ']

def cleaning(review):
  review = re.sub(r'[^\w\s]', '', review)       # íŠ¹ìˆ˜ë¬¸ì ì œê±°
  review = re.sub(r'[\n\t]', ' ', review)       # ì¤„ë°”ê¿ˆ, íƒ­ ì œê±°
  review = re.sub(r'\s+', ' ', review)          # ì—°ì†ëœ ê³µë°± ì œê±°
  review= review.strip()                        # ë¬¸ì¥ ì–‘ë ê³µë°± ì œê±°

  # ë¶ˆìš©ì–´ ì œê±°
  review = ' '.join([w for w in review.split(' ') if w not in stopwords])
  # í•œê¸€ì ì œê±°
  review = ' '.join([w for w in review.split(' ') if len(w) > 1])

  return review

# í´ë Œì§• í•¨ìˆ˜ ì ìš©
df['clean'] = df['text'].apply(cleaning)
df[['text', 'clean']]
```

### í…ìŠ¤íŠ¸ ë­‰ì¹˜ ë°ì´í„°ë¥¼ í…ìŠ¤íŠ¸ íŒŒì¼ë¡œ ìƒì„±

`sentencepiece`ëŠ” í…ìŠ¤íŠ¸ íŒŒì¼ë¡œë¶€í„° í† í°ì„ ì°¾ì•„ë‚´ëŠ” í•™ìŠµì„ ì§„í–‰í•˜ê¸° ë•Œë¬¸ì— ë¨¼ì € ê°€ì§€ê³  ìˆëŠ” í…ìŠ¤íŠ¸ë¥¼ í•˜ë‚˜ì˜ .txtíŒŒì¼ì— ì“°ëŠ” ì‘ì—…ì´ í•„ìš”í•©ë‹ˆë‹¤.

```python
import sentencepiece as spm
import pandas as pd
import re

df = pd.read_csv("nreview_mask.csv")

# ì¶”ê°€ ì“°ê¸°ëª¨ë“œë¡œ í…ìŠ¤íŠ¸ íŒŒì¼ ì—´ê¸°
with open('nreview_mask.txt', 'a', encoding='utf-8') as f:
  for text in df['text']:
        text = str(text)
        text = re.sub(r'[^\w\s]', '', text)  # íŠ¹ìˆ˜ë¬¸ì ì œê±°
        text = re.sub(r'[\n\t]', ' ', text)  # ì¤„ë°”ê¿ˆ, íƒ­ ì œê±°
        text = re.sub(r'\s+', ' ', text)  # ì—°ì†ëœ ê³µë°± ì œê±°
        text= text.strip() # ë¬¸ì¥ ì–‘ë ê³µë°± ì œê±°
        try:
            f.write(text+'\n')
        except:
                pass
```
