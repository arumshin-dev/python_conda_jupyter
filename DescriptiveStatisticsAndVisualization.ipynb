```python
df.describe()#수치형 데이터의 요약정보
```

```python
df.describe(include='all')
# count: 값 개수
# mean: 평균 (수치형에서만)
# std: 표준편차 (수치형에서만)
# min, 25%, 50%, 75%, max: 최소, 1/4, 1/2, 3/4, 최대값 (수치형에서만)
# unique: 고유값 개수 (범주형에서만)
# top: 최빈값 (범주형에서만)
# freq: 최빈값 빈도 (범주형에서만)

# 수치형(Numeric) 데이터의 경우
# 항목	의미
# count	값의 개수 (결측치 제외)
# mean	평균값
# std	표준편차
# min	최소값
# 25%	1사분위수 (하위 25%)
# 50%	중앙값(2사분위수, median)
# 75%	3사분위수 (상위 25%)
# max	최대값
# 범주형(object, string) 데이터의 경우
# 항목	의미
# count	값의 개수 (결측치 제외)
# unique	고유값의 개수
# top	가장 많이 등장한 값(최빈값)
# freq	최빈값의 빈도수
```

```python
import seaborn as sns
df5= pd.read_csv("https://bakey-api.codeit.kr/api/files/resource?root=static&seqId=8305&version=1&directory=bike.csv&name=bike.csv")
df5
```

```python
sns.set_theme(rc={'figure.figsize':(12,6)}, style='white')
sns.barplot(data=df5, x='month', y='total',errorbar=None)
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize':(12,6)}, style='ticks', palette='dark')
sns.barplot(data=df5, x='month', y='total',errorbar=None)
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize':(12,6)}, style='dark')
sns.barplot(data=df5, x='month', y='total',errorbar=None, hue='workingday')
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize':(12,6)}, style='whitegrid')
sns.barplot(data=df5, x='month', y='registered',errorbar=None, hue='workingday')
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize':(12,6)}, style='white')
sns.barplot(data=df5, x='month', y='casual',errorbar=None, hue='workingday')
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize':(12,6)}, style='white')
sns.lineplot(data=df5, x='month', y='casual',errorbar=None, hue='workingday')
plt.show()
```
