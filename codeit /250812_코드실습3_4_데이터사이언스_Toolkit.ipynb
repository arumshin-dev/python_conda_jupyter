```python
# import numpy as np

item = ["비누", "후드티", "청바지", "점퍼", "냄비", "소고기", "커피믹스"]
price = [3700, 54000, 67000, 99000, 89000, 24500, 12000]
quantity = [270, 35, 52, 5, 8, 23, 34]

shop = np.zeros(len(item), dtype=[('상품', 'U10'), ('단가', 'i4'), ('수량', 'i4'), ('매출액', 'i4')])
shop['상품'] = item
shop['단가'] = price
shop['수량'] = quantity
shop['매출액'] = shop['단가'] * shop['수량']

print(shop)
```

```python
np.array(price)*np.array(quantity)
```

```python
# 실습 과제1 - 상품 매출액 계산 답안
price_arr=np.array(price)
quantity_arr=np.array(quantity)

price_arr2=price_arr[1:5]
quantity_arr2=quantity_arr[1:5]

sales = price_arr2*quantity_arr2
print("상품별 매출액: ", sales)
print("전체 매출액: ", sales.sum())
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [25, 30, 20, 25]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode=(0,0.1,0,0)

plt.figure(figsize=(4, 3))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',  # 파이 조각 위에 퍼센트 보여주기
    startangle=90,      # 시작 각도 조정
    explode=explode,    # 특정 조각 강조
    shadow=True,        # 그림자 효과
    counterclock=False, # 시계 방향으로 그리기
    # wedgeprops={'width': 0.5}  # 조각 너비 조정 -도넛 중간에 빔
    )

plt.title('커스터마이징된 파이차트', fontsize=20, fontweight='bold')
plt.show()
```

```python
# 다중 레이아웃 - subplot(행, 열, 순번) 기본

x = np.linspace(0, 2*np.pi, 100)

plt.figure(figsize=(7,3))
plt.subplot(1,2,1)
plt.plot(x, np.sin(x))
plt.title('sin(x) 사인')

plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
plt.title('cos(x) 코사인')

plt.show()
```

```python
# 다중 레이아웃 - subplot(행, 열, 순번) 응용

x = np.linspace(0, 2*np.pi, 100)

plt.figure(figsize=(7,7))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.plot(x, np.sin(x+i), label=f'sin(x+{i})')
    plt.title(f'{i}+1 사인 그래프')
    plt.legend()

plt.tight_layout()
plt.show()
```
