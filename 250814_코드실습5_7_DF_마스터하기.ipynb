# 정규화/표준화

- 정규화 : 변수들끼리 크기 차이가 있을때, 서로 영향을 최소한으로 줄이기 위해 값의 범위를 0~1로 조정하는 것
    - 머신러닝 전처리를 위해
    - 음수 없음

- 표준화 (Z-score Standardization) : 평균을 0, 표준편차를 1로 맞춤
    - 음수 가능
    - (값-평균)/표준편차
    -  x -mean() /std()

```python
# 1. 정규화 - MinMax
# Normalization Norm 놈
from sklearn.preprocessing import MinMaxScaler
# 데이터의 범주를 바꿔줌

#**sales 오른쪽 치우침이 심했던 변수
df['sales_norm'] = (df['sales']-df['sales'].min()) / (df['sales'].max()-df['sales'].min())
df[['sales', 'sales_norm']].head(2)
```

```python
# 머신러닝에서 제일 유명한 라이브러리: 사이킷런
# 위와 같은 코드
# from sklearn.preprocessing import MinMaxScaler # 전처리
scaler = MinMaxScaler()
df['sales_norm'] = scaler.fit_transform(df[['sales']])
df[['sales', 'sales_norm']].head(2)
```

```python
left = pd.DataFrame({
    "id": [1, 2, 3],
    "score": [90, 85, 88]
})

right = pd.DataFrame({
    "id": [2, 3, 4],
    "grade": ["B", "A", "C"]
})
```

```python
print(left)
print('---------')
print(right)
print('---------')
print(left,right)
print('---------')
print(left,f'\n',right)
print('---------')
```

```python
merged = pd.merge(left, right, on='id') # on: 공통컬럼, 합치는 것에 대한 기준
merged
```

```python
merged = pd.merge(left, right, on='id', how='inner') # on: 공통컬럼, 합치는 것에 대한 기준
merged
```

```python
merged = pd.merge(left, right, on='id',how='left') # on: 공통컬럼, 합치는 것에 대한 기준
merged
```

```python
merged = pd.merge(left, right, on='id',how='right') # on: 공통컬럼, 합치는 것에 대한 기준
merged
```

```python
merged_o = pd.merge(left, right, on='id',how='outer') # on: 공통컬럼, 합치는 것에 대한 기준
merged_o
```
